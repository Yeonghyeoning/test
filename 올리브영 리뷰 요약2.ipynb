```python
prompt1 = f"""
다음은 {reviews_df['상품명'][0]} 제품의 간단한 특징입니다.
{item_features_data[0]}

해당 제품의 간단한 특징과 아래 리뷰를 참고하여 제품의 특징에 대해 5문장 내외로 요약해주세요.
"""
```

```python
MY_API_KEY = getpass("OpenAI API Key :")
```

```python
client = OpenAI(api_key=MY_API_KEY)
```

```python
# LLM모델을 호출하여 리뷰들에 대한 요약을 진행하는 사용자 정의 함수 선언
def summarize(reviews, prompt, temperature=0, model='gpt-3.5-turbo'):
    prompt = prompt + "\n\n" + reviews

    completion = client.chat.completions.create(model = model,
                                               messages = [{'role':'user', 'content': prompt}],
                                               temperature = temperature
                                               )
    return completion
```

```python
summarize(reviews_text_data[0][:45000], prompt1, temperature =0, model = 'gpt-4o').choices[0].message.content
```

```python
# 워드클라우드 라이브러리 임포트
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text = reviews_text_data[0]
```

```python
f_path = "C:\Windows\Fonts\gulim.ttc"
wc = WordCloud(font_path=f_path,
               random_state=42, 
               background_color='white',
               stopwords=["리뷰"])

# ====================== 주요 내부 인자들(ctrl + tab으로 더 확인해 보세요~!) ======================
# font_path : 사용할 폰트 경로 및 파일명 지정(C:\Windows\Fonts\폰트명 및 확장자)
# random_state : 랜덤 seed값 설정(설정이 없으면 워드클라우드 생성마다 약간씩 다른 형태로 나올 수 있음)
# max_font_size : 표시되는 가장 큰 텍스트의 크기 제한
# max_word : 최대 단어 개수 제한
# width : 이미지 너비 지정
# height : 이미지 높이 지정
# background_color : 배경색 지정
# stopwords : 불용어 단어 지정(워드클라우드에 출력시키고 싶지 않은 단어를 리스트로 지정)
# ==============================================================================================

# generate : 문자열에서 단어의 빈도(띄워쓰기 기준)를 자동 계산하여 워드클라우드 생성
wc.generate(text)

# 그림의 x,y 좌표값 제거
plt.axis("off")

# interpolation : 보간법(이미 알려진 값 사이의 알려지지 않은 값을 추정하는 것)
#                 출력되는 각 단어 이미지들에 이중선형 보간법 적용(부드러운 이미지로 출력 가능)
plt.imshow(wc, interpolation='bilinear')
plt.show()
```
